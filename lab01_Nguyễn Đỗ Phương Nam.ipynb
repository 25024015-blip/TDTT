### Problem 4
#### Write a function that returns the count of positive integers k such that 1 <= k < n and k > (n - 1) / 2. (Example: n=9 -> 4).

```python
def count_k(n): 
    count = 0
    moc_so_sanh = (n-1)/2
    for k in range (1,n):
        if k > moc_so_sanh:
            count += 1 
    return count
n = count_k(9)
print(n)
```

### Problem 6
#### Write a function that computes the area of a triangle with side lengths a, b, c (use Heron's formula). Raise ValueError if the sides cannot form a triangle.

```python
def dien_tich(c1,c2,c3):
    n = (c1+c2+c3)/2
    kq = (n*(n-c1)*(n-c2)*(n-c3))**1/2
    return kq
print("diện tích hình tam giác là:",dien_tich(2,3,4)) #đây là phần tính diện tích
```

### Problem 8
#### Write a function that returns the character that is n positions after character x in the alphabet.
#### Preserve case and wrap around within A..Z or a..z.

```python
def ki_tu_đoi(x, n): # kiểm tra xem x có phải chữ cái không tránh người dùng nhập số hoặc dấu chấm phẩy
    if x.isalpha():
        # bước 1 xác định bước mốc bắt đầu 
        # kiểm tra xem có phải là chữ in hoa không isupper() 
        if x.isupper():
            base = ord('A') # base sẽ là 65
        else:
            base = ord('a') #base sẽ là 97
        new_position = (ord(x) - base + n) % 26 
        #ord(x) - base là đưa chữ cái về vị trí 0 đến 25
        # + n tức là tiến thêm n bước 
        # % 26 chính là xoay vòng nếu số lớn hơn 26
        new_char_code = new_position + base 
        return chr(new_char_code)        
    else :
        return x
print("Dịch 'a' lên 2 vị trí:", ki_tu_đoi('a', 2))   
print("Dịch 'y' lên 2 vị trí:", ki_tu_đoi('y', 2))   
print("Dịch 'Z' lên 3 vị trí:", ki_tu_đoi('Z', 3))    
```

### Problem 9
#### Write a function that converts an uppercase letter to lowercase. If the input is not a single character, raise ValueError.

```python
def in_thuong(ki_tu):
    # Sửa lại thành dấu gạch dưới _ và số 1
    if len(ki_tu) != 1:
        # Sửa lại đúng chính tả ValueError
        raise ValueError("Đầu vào bắt buộc phải là một kí tự duy nhất")
    return ki_tu.lower()

print("Chuyển 'A' thành:", in_thuong('A'))  # Kết quả: 'a'
print("Chuyển 'Z' thành:", in_thuong('Z'))  # Kết quả: 'z'
```

### Problem 12
#### Write a function that returns a right isosceles triangle of height n using '*'.
#### Line i (1-indexed) has i '*', joined with newlines.
#### Example with n = 4
#### *
#### **
#### ***
#### ****

```python
def ve_tam_giac(n):
    # Bước 1: Tạo một danh sách rỗng để lưu trữ từng dòng của tam giác
    cac_dong = []
    
    # Bước 2: Vòng lặp chạy từ 1 đến n (Lưu ý: range(1, n+1) sẽ dừng ở n)
    for i in range(1, n + 1):
        # Bước 3: Tạo ra chuỗi gồm i dấu sao và thêm nó vào danh sách
        dong_hien_tai = '*' * i
        cac_dong.append(dong_hien_tai)
        
    # Bước 4: Nối tất cả các phần tử trong danh sách bằng ký tự xuống dòng '\n'
    ket_qua = '\n'.join(cac_dong)
    
    # Trả về chuỗi kết quả cuối cùng
    return ket_qua

# --- PHẦN CHẠY THỬ (TEST) ---
# Gắn kết quả của hàm vào một biến
tam_giac_4 = ve_tam_giac(4)

# In thử ra để xem hình thù thế nào
print(tam_giac_4)
```
