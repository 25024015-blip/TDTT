W5A28 Viết hàm tính góc ∠ABC (độ) từ tọa độ A(x1,y1), B(x2,y2), C(x3,y3).


```python
import math
def goc(x1,y1,x2,y2,x3,y3):
    a=(x2-x1)**2+(y2-y1)**2
    b=(x3-x2)**2+(y3-y2)**2
    c=(x3-x1)**2+(y3-y1)**2
    return math.degrees(math.acos((a+b-c)/(2*(a**(1/2))*(b**(1/2))))) 
x1,y1,x2,y2,x3,y3=map(int,input().split())
goc(x1,y1,x2,y2,x3,y3)
                     
```

```python
def nhi_thuc(n,k): # với điều kiện n lớn hơn k
    tich=1 
    tich1=1 
    tich3=1 
    for i in range(1,n+1):
        tich=tich*i
    for j in range(1,k+1):
        tich1=tich1*j 
    for u in range(1,n-k+1):
        tich3=tich3*u
    print(tich/(tich1)*(tich3))       
n,k=map(int,input().split())
nhi_thuc(n,k)    

            
```

Viết hàm sum_1_to_n(n) trả về tổng các số x từ 1 đến n, nếu x có các chữ số đôi một khác nhau


```python
def tong(n): 
    danh_sach=[]
    tong =0
    for i in range(1,n+1):
        ch_so=str(i)
        if len(set(ch_so))==len(ch_so):
            danh_sach.append(i)
    for j in danh_sach:
        tong+=j
    print(tong)
n = int(input())
tong(n)                
```
