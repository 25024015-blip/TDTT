W4A18 Nhập vào số nguyên dương a và b, in toàn bộ ước chung của a và b

```python
a,b = map(int,input().split())
danh_sach = []
danh_sach1 = []
for i in range (1,a+1):
    if a%i==0:
        danh_sach.append(i)
for j in range(1,b+1):
    if b%j==0:
        danh_sach1.append(j)        
chung = set(danh_sach) & set(danh_sach1)
print(sorted(chung))        
```

```python
import math 
danh_sach = []
a,b = map(int,input().split()) 
uocmax = math.gcd(a,b)
for i in range(1,uocmax + 1):
    if uocmax % i ==0:
        danh_sach.append(i)
print(*danh_sach)        
```

W4A25
Nhập vào một dãy số nguyên, ngưng nhập khi người dùng nhập -1.
Sau khi nhập xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập 


```python
n = int(input())
if n!=-1:
    max = n
    min = n
    while n!=-1:
        if n>max:
            max = n
        if n<min:
            min = n
        n = int(input())
    print(f"{max} {min}")    
else:
    print("chưa có số nào để kiểm tra")                
```

W4A26
[Fibonaci1]
Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số kế tiếp sẽ
bằng tổng hai số trước đó
Nhập vào số nguyên dương A, hãy tìm số trong dãy số fibonacci lớn nhất nhưng không vượt
quá A 


```python
A = int(input())
f1 = 1
f2 = 1
while f2<=A:
    f1,f2=f2,f2+f1
print(f1)    
```

thường và thường thành hoa
