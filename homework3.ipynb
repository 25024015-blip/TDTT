Bài 8: Nhập vào số nguyên dương a và b, in toàn bộ ước dương chung của a và b 

```python
a,b = map(int,input().split())
danh_sach=[]
danh_sach1 =[]
for i in range(1,a+1):
    if a%i==0:
        danh_sach.append(i)
for j in range(1,b+1):
    if b%j==0:
        danh_sach1.append(j)
chung = set(danh_sach) & set(danh_sach1)# chỉ có kiểu dữ liệu set mới dùng phép giao được
print(sorted(list(chung)))
```

```python
# cách 2 vốn dĩ toàn bộ ước chung của a và b thì đều là ước của ước chung lớn nhất của a và b
import math 
a,b = map(int,input().split())
uoc_chung = math.gcd(a,b)
danh_sach = []
for i in range(1,uoc_chung +1):
    if uoc_chung%i==0:
        danh_sach.append(i)
print(*danh_sach)        
```

Bài 10: Nhập vào số nguyên dương a, kiểm tra xem a có phải là số hoàn hảo hay không.
Số hoàn hảo là số có tổng các ước (ước nhỏ hơn nó) bằng chính nó, VD: số 6, 28… 

```python
a = int(input())
if a>0:
    tong = 0
    for i in range(1,a):
        if a%i==0:
            tong+=i
    if tong==a:
        print("là số hoàn hảo")  
else:
    print("không thỏa mãn điều kiện")              
```

Bài 12: Nhập số nguyên dương n, Cho S(k) = 1 + 2 + 3 + … + k. Tìm k sao cho S(k)
lớn nhất nhưng nhỏ hơn n 

```python
n = int(input())
s=0
k=0
while s+(k+1)<n: # cố gắng dùng while nhiều hơn
    k+=1
    s+=k
print(k)    
```

Bài 14: Nhập vào một dãy số nguyên, dừng nhập khi người dùng nhập -1. Sau khi nhập
xong, in số lớn nhất, số nhỏ nhất trong những số vừa nhập. KHÔNG ĐƯỢC dùng mảng

```python
n = int(input())
if n!=-1:
    max = n
    min = n # đặt mốc 
    while n!=-1:
        if n>max:
            max = n
        if n<min:
            min = n
        n = int(input()) 
    print(f"{max} {min}")
else:
    print("chưa có số nào để kiểm tra")




```

Bài 18: Nhập vào một số nguyên dương n, kiểm tra xem n có phải là số dạng 3^k hay

Bài 19: Dãy số fibonacci là dãy số được định nghĩa như sau: 1, 1, 2, 3, 5, 8, 13,... với số
kế tiếp sẽ bằng tổng hai số trước đó. Nhập vào số nguyên dương A, hãy tìm số trong
dãy số fibonacci lớn nhất nhưng không vượt quá A 

```python
a = int(input())
f1 = 1
f2 = 1
while f2<=a:
    f1,f2=f2,f2+f1
print(f1)    
```

Bài 20: Nhập vào số nguyên dương n, đếm xem n có bao nhiêu chữ số là số nguyên tố. 

```python
n = int(input())
chuoi = str(n)
dem1 = 0
for i in chuoi:
    h = int(i)
    dem = 0
    for j in range(1,h+1):
        if h%j==0:
            dem+=1
    if dem == 2:   
        dem1+=1
print(dem1)              
```
