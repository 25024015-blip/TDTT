# Bài 2: Một khu đất hình chữ nhật có chiều rộng a, chiều dài b. Người ra xây một khu
vui chơi hình tròn tại vị trí trung tâm khu đất như hình vẽ. Nhập vào a, b. Hãy tính phần
diện tích còn để trồng cây xung quanh là bao nhiêu. Biết pi = 3.14, lấy kết quả đến 2
chữ số phần thập phân. 

```python
a =float(input("nhập chiều rộng a:")) 
b = float(input("hãy nhập chiều dài b:"))
hcn = a*b
bk = a/2
ht = 3.14*(bk**2) 
cay = hcn - ht
print(f"diện tích còn lại để trồng cây xung quanh là {cay:.2f}")    
```

# Bài 6. Nhập vào 3 số a, b, c. Kiểm tra xem đây có phải là cạnh của 1 tam giác không,
nếu đúng thì tính diện tích của tam giác đó, in ra với 1 chữ số phần thập phân, nếu nó
không phải 3 cạnh của tam giác thì in ra “Khong phai 3 canh cua tam giac”. 

```python
a = int(input("nhập vào cạnh a"))
b = int(input("hãy nhập vào cạnh b")) 
c = int(input("hãy nhập vào cạnh c")) 
if (a+b>c and a+c>b and c+b>a):
    h = (a+b+c)/2
    s = (h*(h-a)*(h-b)*(h-c))**(1/2)
    print("diện tích hình tam giác là", s)
else:
    print("không phải là ba cạnh của tam giác")
```

9. Chuyển đổi số thành chữ số: Nhập một số nguyên từ 0 đến 9, in ra tên chữ số
tương ứng (0 -> "Không", 1 -> "Một", ...) – Xem trước match …case

```python
n = int(input()) 
if n==0 :
    print("Không")
if n == 1:
    print("Một") 
if n == 2:   
    print("Hai") 
if n == 3:
    print("Ba") 
if n == 4:
    print("Bốn") 
if n == 5 :
    print("Năm") 
if n == 6:  
    print("Sáu") 
if n == 7: 
    print("Bảy") 
if n == 8:
    print("Tám ") 
if n == 9:
    print("Chín") 

            
```

```python
n = int(input()) 
match n :
    case 0:print("Không")
    case 1 :print("Một") 
    case 2:print("Hai") 
    case 3:print("Ba")
    case 4:print("Bốn") 
    case 5:print("Năm") 
    case 6:print("Sáu") 
    case 7:print("Bảy") 
    case 8:print("Tám ") 
    case 9:print("Chín") 
    
```
