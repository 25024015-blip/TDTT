W4A10
"[LongestCollatzUpToN]
Nhập vào số nguyên dương n. Tìm số x trong [1, n] có độ dài dãy Collatz lớn nhất
và in ra x cùng độ dài dãy tương ứng (x L). Nếu có nhiều số cùng độ dài, chọn số
nhỏ nhất." 


```python
n = int(input())
max_len = 0
best_x = 1
for i in range(1, n + 1):
    temp = i      
    do_dai = 1     
    while temp != 1:
        if temp % 2 == 0:
            temp = temp // 2  
        else:
            temp = 3 * temp + 1
        do_dai += 1
    if do_dai > max_len:
        max_len = do_dai
        best_x = i
print(best_x, max_len)
```

Nhập một số nguyên N(N,10000).in ra tất cả các số nguyên tố <=N.Dùng sàng lọc Erastosthen

```python
N = 10000 #đầu tiên thì đây chỉ là giới hạn
dd = [True]*(N+1) # ta gán để sử dụng mảng tại ban đầu ta sẽ cho mọi số từ 0 đến N+1 đều là số nguyên tố
for i in range(2,N+1):
    if dd[i]==True:
        for j in range(i*2,N,i):
            dd[j]=False
for i in range(2,N+1):
    if dd[i]==True:
        print(i,end=" ")            
```
