W3A3
Kiểm tra xem một số có phải là lũy thừa của 2 hay không
Sử dụng các toán tử bitwise để xác định xem một số có phải là lũy thừa của 2 hay không.
Gợi ý: Đối với lũy thừa của 2, n & (n-1) == 0 

```python
n = int(input()) 
if n > 0 and(n & (n-1))==0 :
    print("là lũy thừa của 2") 
else :   
    print("không là lũy thừa của 2") 
```

W3A5 viêt chương trình chia một số m cho n, kết quả thu được làm tròn lên 


```python
m,n = map(float,input().split()) 
k = int(input())
kq = m/n 
kq1 = kq*(10**(k+1))
kq2 = int(kq1)
dk = kq2%10
base = kq*10**k 
base2 = int(base) 
if dk >=5:
    base2 = base2 + 1
base1 = base2/10**k
print(base1)    
```

```python
a,b,c = map(int,input().split()) 
if a+b>c and a+c>b and b+c>a :
    if a==b==c:
        print("Tam giác đều")
    elif a==b or a==c or b==c :
        print("tam giác cân") 
    else :
        print("đây là tam giác thường")    
else:
    print("không phải  tam giác")        
            
```
