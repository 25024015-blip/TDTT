```python
import math

def smart_taxi_fare(car_type, start_hour, km, wait_minutes):
    # Khai báo giờ phụ thu
    hphu = [22, 23, 0, 1, 2, 3, 4, 5]
    
    tien_xe = 0
    tien_cho = 0
    
    # 1 & 2. Tính tiền xe theo km
    if car_type == "xe 4 chỗ":
        if km <= 0.8:
            tien_xe = 11000
        elif km <= 20:
            tien_xe = 11000 + (km - 0.8) * 13000
        else:
            tien_xe = 11000 + 19.2 * 13000 + (km - 20) * 10500
            
    elif car_type == "xe 7 chỗ":
        if km <= 0.8:
            tien_xe = 12000
        elif km <= 20:
            tien_xe = 12000 + (km - 0.8) * 14000
        else:
            tien_xe = 12000 + 19.2 * 14000 + (km - 20) * 12000
            
    # 3. Tính tiền chờ
    if wait_minutes <= 5:
        tien_cho = 0
    else:
        # Làm tròn lên số block 5 phút (ví dụ 6 phút -> 1 block)
        so_block = math.ceil((wait_minutes - 5) / 5)
        tien_cho = so_block * 2000
        
    # Tính tổng tạm
    total = tien_xe + tien_cho
    
    # 4. Phụ thu ban đêm
    if start_hour in hphu:
        total = total * 1.2  # Cộng 20%
        
    # 5. Điều kiện đặc biệt
    if km > 30 and start_hour not in hphu:
        total = total * 0.9  # Giảm 10%
        
    if total < 50000 and km > 5:
        total = 50000
        
    # Yêu cầu trả về số nguyên
    return int(total)

# Cách gọi hàm đúng (không để input trong hàm)
# Bạn có thể test hàm bằng cách truyền giá trị vào như sau:
ket_qua = smart_taxi_fare("xe 4 chỗ", 0, 5000, 5000)
print("Tổng tiền:", ket_qua)
```

```python
#cách 2 đây là cách mà em copy trên gemini để kiểm tra đáp án
import math


def goc(x1, y1, x2, y2, x3, y3):
    # Tính bình phương độ dài các cạnh
    ABbinh = (x2 - x1)**2 + (y2 - y1)**2
    ACbinh = (x3 - x1)**2 + (y3 - y1)**2
    BCbinh = (x3 - x2)**2 + (y3 - y2)**2  # Đã sửa lại tọa độ của B
    
    # Xử lý trường hợp 2 điểm trùng nhau (độ dài bằng 0) thì không tạo thành góc
    if ABbinh == 0 or BCbinh == 0:
        return "không hợp lệ"
        
    # Định lý hàm số cosin
    cosb = (ABbinh + BCbinh - ACbinh) / (2 * math.sqrt(ABbinh) * math.sqrt(BCbinh))
    
    # Chặn sai số dấu phẩy động để đảm bảo cosb luôn nằm trong khoảng [-1, 1]
    cosb = max(-1.0, min(1.0, cosb))
    
    # Tính góc bằng radian rồi đổi sang độ
    kq_radian = math.acos(cosb)
    kq_do = math.degrees(kq_radian)
    
    return kq_do # Trả về giá trị theo đúng yêu cầu đề

# --- Phần chạy thử nghiệm ---
x1, y1, x2, y2, x3, y3 = map(int, input().split())
ket_qua = goc(x1, y1, x2, y2, x3, y3)
print(ket_qua)
```

30 Yêu cầu: Viết hàm detect_number_type(s) trả về 'int' nếu s biểu diễn một số nguyên, 'float' nếu biểu diễn một số thực dấu
chấm động, ngược lại trả về 'neither'.
- Đầu vào: chuỗi s.
- Đầu ra: một trong các giá trị {'int', 'float', 'neither'}.
- Quy tắc:
- Cho phép có dấu + hoặc - ở đầu.
- Số nguyên phải chỉ gồm các chữ số và không có dấu chấm thập phân (ví dụ: '100').
- Số thực phải có dấu chấm và có chữ số ở cả hai phía của dấu chấm (ví dụ: '100.0', '-3.14').
- Các chuỗi như '.5', '5.', '1e3' được xem là 'neither' trong bài này. 

```python
def kiem_tra(s):
    if s=="":
        return "neither"
    if s[0]=="+" or s[0]=="-":
        s=s[1:]
    if s=="":
        return "neither"
    if s.isdigit():
        return "int"
    if s.count(".")==1:
        cat=s.split(".")
        dau=cat[0]
        duoi=cat[1]
        if dau.isdigit()==True and duoi.isdigit()==True:
            return "float"
    return "neither" 
s=input()
print(kiem_tra(s))   
```

35 Yêu cầu: Viết hàm divide_with_remainder(dividend, divisor) trả về bộ (q, r) trong đó q là thương nguyên và r là số dư sao
cho dividend = divisorq + r với 0 <= |r| < |divisor| (dùng ngữ nghĩa của Python với // và %).
- Đầu vào: số nguyên dividend, divisor (divisor khác 0).
- Đầu ra: bộ hai số nguyên (q, r).
- Ví dụ: divide_with_remainder(5, 3) -> (1, 2). 


```python
def divide_with_remainder(dividend, divisor): 
    q=dividend//divisor 
    r=dividend-divisor 
    if divisor !=0:
        for i in range(-divisor+1,divisor):  
            if i==r:
                return f"({q},{r})"
    else:
        raise ValueError 
dividend, divisor=map(int,input().split())
print( divide_with_remainder(dividend, divisor))   
         
```

50.chuỗi hạnh phúc chỉ gồm các kí tự hoa ,thường hoặc dấu cách và có mặt đủ từ "a" đến "z" ,không kể hoa hay thường

```python
def chuoi_hanh_phuc(s):
    # 1. Kiểm tra điều kiện: CHỈ gồm kí tự hoa, thường hoặc dấu cách
    for char in s:
        # Nếu kí tự không phải là chữ cái (isalpha) VÀ không phải dấu cách
        if not (char.isalpha() or char == ' '):
            return False # Phát hiện "kẻ xâm nhập", trả về False luôn
            
    # 2. Kiểm tra điều kiện: Có mặt ĐỦ từ 'a' đến 'z'
    s_thuong = s.lower() # Đưa hết về chữ thường để dễ so sánh
    bang_chu_cai = "abcdefghijklmnopqrstuvwxyz"
    
    for chu_cai in bang_chu_cai:
        if chu_cai not in s_thuong:
            return False # Thiếu dù chỉ 1 chữ cái cũng trả về False
            
    # Vượt qua cả 2 bài kiểm tra khắc nghiệt thì đây là chuỗi hạnh phúc
    return True

# --- PHẦN CHẠY THỬ (TEST) ---
chuoi_1 = "The quick brown fox jumps over the lazy dog"
chuoi_2 = "The quick brown fox jumps over the lazy dog 123" # Bị vướng số
chuoi_3 = "Xin chao the gioi" # Không đủ 26 chữ cái

print(f"Chuỗi 1: {chuoi_hanh_phuc(chuoi_1)}") # Mong đợi: True
print(f"Chuỗi 2: {chuoi_hanh_phuc(chuoi_2)}") # Mong đợi: False
print(f"Chuỗi 3: {chuoi_hanh_phuc(chuoi_3)}") # Mong đợi: False
```
