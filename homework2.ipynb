```python
a,b,c = map(float,input().split())
if a+b>c and a+c>b and b+c>a: #sử dụng if lồng nhau
    if a==b==c: 
        print("đây là tam giác đều") 
    elif a==b>c or a==b<c and a==c>b or a==c<b and b==c>a or b==c<a : 
         print("đây là tam giác cân") 
    elif (a**2+b**2)**(1/2)== c or (a**2+c**2)**(1/2)==b or  (b**2+c**2)**(1/2)==a: 
        print("đây là tam giác vuông ") 
    n = (a+b+c)/2 
    print(f"diện tích là  {(n*(n-a)*(n-b)*(n-c))**(1/2)} \n chu vi là {a+b+c}") 
else:
    print("đây không phải ba cạnh tam giác ")    
       
    
```

Bài 8: Giải hệ phương trình tuyến tính:
ax  by  m
cx  dy  n

```python
a,b,m,c,d,n = map(float,input().split())
det = a*d-c*b 
det_x= m*d-b*n 
det_y = a*n-n*m 
if det == 0: 
    if det == 0 and det_x ==0 and det_y ==0 :
        print("hệ có vô số nghiệm")
    elif det == 0 and (det_x != 0 or det_y != 0) :
        print("hệ vô nghiệm ") 
else:
    print(f"hai nghiệm của phương trình là {(det_x,det_y)}")            



   
```
